```python
from PIL import Image
import numpy as np
import pandas as pd
from skimage import color
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

def kmeanplot(target, k) :
    ms = KMeans(n_clusters=k).fit(target)
    labels = ms.labels_
    cluster_centers = ms.cluster_centers_

    labels_unique = np.unique(labels)
    n_clusters_ = len(labels_unique)

    print("number of estimated clusters : %d" % n_clusters_)

    # #############################################################################
    # Plot result
    import matplotlib.pyplot as plt
    from itertools import cycle
    
    fig = plt.figure(figsize=(12, 6))
    ax = plt.axes(projection='3d')
    colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
    for k, col in zip(range(n_clusters_), colors):
        my_members = labels == k
        cluster_center = cluster_centers[k]
        ax.scatter(target[my_members, 0], target[my_members, 1], target[my_members, 2])
        #plt.plot(target[my_members, 0], target[my_members, 1], target[my_members, 2], col + '.')
        #plt.plot(cluster_center[0], cluster_center[1], cluster_center[2], 'o', markerfacecolor=col,
        #        markeredgecolor='k', markersize=14)
    plt.title('Estimated number of clusters: %d' % n_clusters_)
    plt.show()
    return ms

MX = [1,5,-4,-3,9,11]
MY = [1,7,-10,6,0,11]
MZ = [1,-3,5,-1,0,11]
SX = [2,2.5,3,2.7,1,1.5]
SY = [1,3,2,1.5,1.5,2.5]
SZ = [1.5,2,1,3,2,3.5]
n_samples = 300
test_n_samples = 100
train_samples = []
test_samples = []
real = []
predict = []

for k in range (0,5) :
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(111, projection='3d')

    sx = np.random.normal(MX[k], SX[k], n_samples)
    sy = np.random.normal(MY[k], SY[k], n_samples) 
    sz = np.random.normal(MZ[k], SZ[k], n_samples)
    train_samples.append([ [sx[i],sy[i],sz[i]] for i in range(0,n_samples)])
    
    ax.scatter(sx, sy, sz, c = sz, s= 20, alpha=0.5, cmap=plt.cm.Greens)
    plt.title(f'Class = {k}')
    plt.show()

target = np.array([train_samples]).reshape((-1,3))
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter([target[i][0] for i in range (0,1500)],[target[i][1] for i in range (0,1500)],[target[i][2] for i in range (0,1500)], c = [target[i][2] for i in range (0,1500)], s= 20, alpha=0.5, cmap=plt.cm.Greens)
ax.scatter([target[i][0] for i in range (0,1500)],[target[i][1] for i in range (0,1500)],[target[i][2] for i in range (0,1500)], c = [target[i][2] for i in range (0,1500)], s= 20, alpha=0.5, cmap=plt.cm.Greens)
ax.scatter([target[i][0] for i in range (0,1500)],[target[i][1] for i in range (0,1500)],[target[i][2] for i in range (0,1500)], c = [target[i][2] for i in range (0,1500)], s= 20, alpha=0.5, cmap=plt.cm.Greens)
ax.scatter([target[i][0] for i in range (0,1500)],[target[i][1] for i in range (0,1500)],[target[i][2] for i in range (0,1500)], c = [target[i][2] for i in range (0,1500)], s= 20, alpha=0.5, cmap=plt.cm.Greens)
ax.scatter([target[i][0] for i in range (0,1500)],[target[i][1] for i in range (0,1500)],[target[i][2] for i in range (0,1500)], c = [target[i][2] for i in range (0,1500)], s= 20, alpha=0.5, cmap=plt.cm.Greens)
plt.title(f'All training sample point')
plt.show()

for k in range (0,6) :
    sx = np.random.normal(MX[k], SX[k], n_samples)
    sy = np.random.normal(MY[k], SY[k], n_samples)
    sz = np.random.normal(MZ[k], SZ[k], n_samples)
    test_samples.append([ [sx[i],sy[i],sz[i]] for i in range(0,test_n_samples)])
    real.append([k for i in range(0,test_n_samples)])
real = np.array(real).flatten()
```

```python
target = np.array([train_samples]).reshape((-1,3))
ms = kmeanplot(target,5)
```

```python
predict = ms.predict(np.array([test_samples]).reshape((-1,3)))
```

```python
arr = [[],[],[],[],[]]
for i in range(0,1500) :
    arr[label.index(ms.labels_[i])].append(i)
```

```python
cov = []
for k in range(0,5) :
    t = np.empty((0,3), float)  
    for i in arr[k] :
        t = np.append(t, np.array([np.array([train_samples]).reshape((-1,3))[i]]), axis = 0)
    a = np.cov(t, rowvar=False)
    cov.append(a)
print(label.index(ms.labels_[arr[4][1]]))
```

```python
tcount = [0,0,0,0,0,0]
fcount = [0,0,0,0,0,0]
tests = np.array([test_samples]).reshape((-1,3))
for i in range (0,test_n_samples*6) :
    x = ((ms.cluster_centers_[predict[i]][0] - tests[i][0])**2)/cov[label.index(predict[i])][0][0]
    y = ((ms.cluster_centers_[predict[i]][1] - tests[i][1])**2)/cov[label.index(predict[i])][1][1]
    z = ((ms.cluster_centers_[predict[i]][2] - tests[i][2])**2)/cov[label.index(predict[i])][2][2]
    dist = x+y+z
    if real[i] == label.index(predict[i]) and dist < 20:
        tcount[real[i]] = tcount[real[i]] + 1
    elif real[i] == 5 and dist > 20 :
        tcount[5] = tcount[5] + 1
    else :
        fcount[real[i]] = fcount[real[i]] + 1
print(f'tcount : {tcount} fcount : {fcount}')
```
